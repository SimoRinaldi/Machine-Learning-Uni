### Plot in pie charts the number of "tamponi_test_molecolare" with percentage for each region with exploded slice for *Lombardia* region

`hint: df.index, df[].plot.pie, figsize=, autopct="%1.1f%%, pctdistance=, explode=, ylabel=", `

- Figure size of (8,8)
- Explode of 0.1 for Lombardia region
- Distance of the percentage of 0.75

```python
explode = [0.1 if region == 'Lombardia' else 0 for region in df.index]

df['tamponi_test_molecolare'].plot.pie(
    figsize=(8, 8),
    autopct="%1.1f%%",
    pctdistance=0.75,
    explode=explode,
    ylabel=""
)
```

### Plot hexbin plot of "deceduti" by ("long", "lat) coordinates of regions

`hint: df.plot.hexbin, x=, y=, C=, edgecolor=, gridsize=`

- Figure size of (8, 6)
- Gridsize of 10

```python
df.plot.hexbin(
    x='long',
    y='lat',
    C='deceduti',
    edgecolor='white',
    gridsize=10,
    figsize=(8, 6)
)
```

### Plot scatterplots of "terapia_intensiva" and "ingressi_terapia_intensiva" by ("long", "lat") coordinates of regions both in the same plot

`hint: df.plot.scatter, x=, y=, color=, alpha=, s=, label=, ax=, ax.legend`

- Figure size of (8, 6)
- Alpha of 0.5
- Legend at the bottom

```python
ax = df.plot.scatter(x='long', y='lat', color='blue', alpha=0.5, s=df['terapia_intensiva'], label='terapia_intensiva', figsize=(8, 6))
df.plot.scatter(x='long', y='lat', color='red', alpha=0.5, s=df['ingressi_terapia_intensiva'], label='ingressi_terapia_intensiva', ax=ax)
ax.legend(loc='lower center')
```

```python
regions_colors = {
    "Abruzzo": "orange", "Basilicata": "orange", "Calabria": "orange",
    "Campania": "red", "Emilia-Romagna": "red", "Friuli Venezia Giulia": "red",
    "Lazio": "red", "Liguria": "orange", "Lombardia": "red", "Marche": "red",
    "Molise": "red", "P.A. Bolzano": "orange", "P.A. Trento": "red",
    "Piemonte": "red", "Puglia": "red", "Sardegna": "white", "Sicilia": "red",
    "Toscana": "orange", "Umbria": "orange", "Valle d'Aosta": "orange", "Veneto": "red"
}

colors = pd.Series(regions_colors)

df.groupby(colors)['nuovi_positivi'].mean().plot.bar(
    figsize=(8, 6),
    title='Mean nuovi_positivi by region color',
    rot=0
)
```

### Group the regions by color and plot bars of mean *and error* of "nuovi_positivi" column

`hint: df.groupby().std(), yerr=, rot=,`

```python
grouped = df.groupby(colors)['nuovi_positivi']

grouped.mean().plot.bar(
    figsize=(8, 6),
    yerr=grouped.std(),
    title='Mean and error of nuovi_positivi by region color',
    rot=0
)
```

### Group the regions by color and plot a hist of "variazione_totale_positivi" using 5 bins

`hint: pd.Series, df.groupby().plot.hist(), legend=,`

```python
colors = pd.Series(regions_colors)

df.groupby(colors)['variazione_totale_positivi'].plot.hist(
    bins=5,
    alpha=0.5,
    legend=True,
    figsize=(8, 6)
)
```
