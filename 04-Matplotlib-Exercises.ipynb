### Plot a line with formula $y = 2x + 1$
`hint: np.arange, plt.figure, plt.plot, plt.title, plt.xlabel, plt.ylabel, plt.grid, plt.legend, plt.show`

- Use points in range [1, 20]
- Figure size of (10,5)
- Set axis labels
- Set plot title
- Set line color as red
- Discontinued line (--) with star (\*) on point
- Plot the legend with the formula in latex version (`r"$ ... $"`)
- Set a dashed grid

```python
x = np.arange(1,21,1)
plt.figure(figsize=(10,5))
plt.plot(x, 2*x + 1, 'r--', label="$y=2x+1$")
plt.plot(x, 2*x + 1, '*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot title')
plt.grid(True, linestyle='--')
plt.legend(loc=2)
plt.show()
```

### Add to the previous plot the line with formula $y = log(x) + 1$ in the same range
`hint: np.log, plt.plot`

- Set line color as blue
- Normal line (-) with square (s) char on point

```python
plt.figure(figsize=(10,5))
plt.plot(x, 2*x+1, 'r*--', label="$y=2x+1$")
y = np.log(x)+1
plt.plot(x, y, 'bs-', label="$y=log(x)+1$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot title')
plt.grid(True, linestyle='--')
plt.legend(loc=2)
plt.show()
```

### Add to the previous plot the line with formula $y=x^2 - 1$ in the same range
`hint: x**2, plt.plot`

- Set line color as green
- Normal line (-) with + (plus) char on point

```python
plt.figure(figsize=(10,5))
plt.plot(x, 2*x+1, 'r*--', label="$y=2x+1$")
y = np.log(x)+1
plt.plot(x, y, 'bs-', label="$y=log(x)+1$")
y = x**2-1
plt.plot(x, y, 'g+-', label="$y=x^2-1$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot title')
plt.grid(True, linestyle='--')
plt.legend(loc=2)
plt.show()
```

### Replot the previous plot in different subplots in the same line without sharing axes
`hint: plt.subplots, axes[].plot, axes[].set_title, axes[].set_xlabel, axes[].set_ylabel, axes[].grid, axes[].legend, fig.tight_layout`

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,5))

y = 2*x+1
axes[0].plot(x, y, 'r*--', label="$y=2x+1$")
axes[0].set_title('Plot 1')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].grid(True, linestyle='--')
axes[0].legend(loc=2)

y = np.log(x)+1
axes[1].plot(x, y, 'bs-', label="$y=log(x)+1$")
axes[1].set_title('Plot 2')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].grid(True, linestyle='--')
axes[1].legend(loc=2)

y = x**2-1
axes[2].plot(x, y, 'g+-', label="$y=x^2-1$")
axes[2].set_title('Plot 3')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].grid(True, linestyle='--')
axes[2].legend(loc=2)

fig.tight_layout()
```

### Replot the previous plot sharing y between suplots
`hint: sharey=True`

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,5), sharey=True)

y = 2*x+1
axes[0].plot(x, y, 'r*--', label="$y=2x+1$")
axes[0].set_title('Plot 1')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].grid(True, linestyle='--')
axes[0].legend(loc=2)

y = np.log(x)+1
axes[1].plot(x, y, 'bs-', label="$y=log(x)+1$")
axes[1].set_title('Plot 2')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].grid(True, linestyle='--')
axes[1].legend(loc=2)

y = x**2-1
axes[2].plot(x, y, 'g+-', label="$y=x^2-1$")
axes[2].set_title('Plot 3')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].grid(True, linestyle='--')
axes[2].legend(loc=2)

fig.tight_layout()
```

### Replot the previous plot using the for loop, lists and dictionaries
`hint: functions = [...], {"title": ..., "y": ..., "label": ..., "linestyle": ...}, zip(functions, axes)`

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,5), sharey=True)

plot1 = {
    'title': 'Plot 1',
    'y': 2*x+1,
    'label': '$y=2x+1$',
    'linestyle': 'r*--'
}

plot2 = {
    'title': 'Plot 2',
    'y': np.log(x)+1,
    'label': '$y=log(x)+1$',
    'linestyle': 'bs-'
}

plot3 = {
    'title': 'Plot 3',
    'y': x**2-1,
    'label': '$y=x^2-1$',
    'linestyle': 'g+-'
}

functions = [plot1, plot2, plot3]

for (func, ax) in zip(functions, axes):
  ax.plot(x, func['y'], func['linestyle'], label=func['label'])
  ax.set_title(func['title'])
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.grid(True, linestyle='--')
  ax.legend(loc=2)

fig.tight_layout()
```

### Plot functions $ y_i = i\cdot x - 5i$ with $i$ in range (0, 10) in the same plot with size (10,5) and a different color for each function

```python
i_range = np.arange(11)

plt.figure(figsize=(10,5))

for i in i_range:
  y = i*x - 5*i
  plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Functions')
plt.grid(True, linestyle='--')
```

### Plot functions $ y_i = i\cdot x - 5i$ with $i$ in range (0, 10) in different subplots with shared y and figure size (20,5)

```python
i_range = np.arange(11)

fig, axes = plt.subplots(nrows=1, ncols=11, figsize=(20,5), sharey=True)

for (i, ax) in zip(i_range, axes):
  y = i*x - 5*i
  ax.plot(x, y, label=f'$y={i}*x-5*{i}')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.grid(True, linestyle='--')
  ax.legend(loc=2)

fig.tight_layout()
```

### Plot the area between $x$ squared and the $x \cdot log(x)$ function in green and alpha $0.5$
`hint: plt.fill_between`

```python
y1 = x**2
y2 = x*np.log(x)
plt.fill_between(x, y1, y2, color='green', alpha=0.5)
plt.title('area between x^2 and x*log(x)')
```

### Replot previous plots in square grid (2,2)
`hint: axes[][]`

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

y1 = x**2
y2 = x*np.log(x)

axes[0][0].scatter(x, y1, color='red')
axes[0][0].set_title('scatterplot x^2')

axes[0][1].step(x, y1, color='blue', linewidth=3)
axes[0][1].set_title('step plot x^2')

axes[1][0].bar(x, y1, color='orange', alpha=0.25)
axes[1][0].set_title('barplot x^2')

axes[1][1].fill_between(x, y1, y2, color='green', alpha=0.5)
axes[1][1].set_title('area between x^2 and x*log(x)')

fig.tight_layout()
```
