### Plot a line with formula $y = 2x + 1$
`hint: np.arange, plt.figure, plt.plot, plt.title, plt.xlabel, plt.ylabel, plt.grid, plt.legend, plt.show`

- Use points in range [1, 20]
- Figure size of (10,5)
- Set axis labels
- Set plot title
- Set line color as red
- Discontinued line (--) with star (\*) on point
- Plot the legend with the formula in latex version (`r"$ ... $"`)
- Set a dashed grid

```python
points = np.arange(1,21,1)
plt.figure(figsize=(10,5))
plt.plot(points, 2*points + 1, 'r--', label="$y=2x+1$")
plt.plot(points, 2*points + 1, '*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot title')
plt.grid(True, linestyle='--')
plt.legend(loc=2)
plt.show()
```

### Add to the previous plot the line with formula $y=x^2 - 1$ in the same range
`hint: x**2, plt.plot`

- Set line color as green
- Normal line (-) with + (plus) char on point

```python
plt.figure(figsize=(10,5))
plt.plot(x, 2*x+1, 'r*--', label="$y=2x+1$")
y = np.log(x)+1
plt.plot(x, y, 'bs-', label="$y=log(x)+1$")
y = x**2-1
plt.plot(x, y, 'g+-', label="$y=x^2-1$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot title')
plt.grid(True, linestyle='--')
plt.legend(loc=2)
plt.show()
```

### Replot the previous plot in different subplots in the same line without sharing axes
`hint: plt.subplots, axes[].plot, axes[].set_title, axes[].set_xlabel, axes[].set_ylabel, axes[].grid, axes[].legend, fig.tight_layout`

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,5))

y = 2*x+1
axes[0].plot(x, y, 'r*--', label="$y=2x+1$")
axes[0].set_title('Plot 1')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].grid(True, linestyle='--')
axes[0].legend(loc=2)

y = np.log(x)+1
axes[1].plot(x, y, 'bs-', label="$y=log(x)+1$")
axes[1].set_title('Plot 2')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].grid(True, linestyle='--')
axes[1].legend(loc=2)

y = x**2-1
axes[2].plot(x, y, 'g+-', label="$y=x^2-1$")
axes[2].set_title('Plot 3')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].grid(True, linestyle='--')
axes[2].legend(loc=2)

fig.tight_layout()
```
